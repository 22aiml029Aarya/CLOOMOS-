```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel(r"C:\Users\pravi\Cloomos\Data_set.xlsx")
data.head()
```

```python
data = data.drop(columns =['Sr. no.','Name','Surname','email address','mobile no.'])
```

```python
data['Type of t-shirt'].unique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_obj = TfidfVectorizer()
tfidf_matrix = tfidf_obj.fit_transform(data['Description'])
tfidf_data = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_obj.get_feature_names_out())
tfidf_data.head()
```

```python
len(tfidf_data.columns.unique())
```

```python
data =pd.concat([data,tfidf_data],axis='columns')
data.head()
```

```python
data = data.drop(columns =['Description'])
ohe_features = ['Type of t-shirt', 'Color','Gender','Location']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe =OneHotEncoder()
encoded_data =ohe.fit_transform(data[ohe_features]).toarray()
ohe_df =pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(ohe_features))
ohe_df.head()
```

```python
data=pd.concat([data,ohe_df],axis='columns')
data.head()
```

```python
day=[]
month=[]
Year=[]
for item in list(data.Date):
    day.append(item.split('/')[0])
    month.append(item.split('/')[1])
    Year.append(item.split('/')[2])

raw_df ={"Day":day,"Month":month,"Year":Year}
date_df = pd.DataFrame(raw_df)
date_df.head()
```

```python
data= data.drop(columns=['Date','Location','Gender','Type of t-shirt'])
data = pd.concat([data,date_df],axis="columns")
data.head()
```

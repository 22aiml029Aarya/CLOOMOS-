```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel(r'Data_set.xlsx')
data
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data = data.drop_duplicates()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
```

```python
sns.histplot(data['Age'], bins=10)
plt.show()
```

```python
sns.countplot(x='Gender', data=data)
plt.show()
```

```python
sns.boxplot(x='Type of t-shirt', y='Price', hue='Color', data=data)
plt.show()
```

```python
sns.histplot(data['Ratings'], bins=10)
plt.show()
```
